```
import sys
import math
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs

#get euler distance between two given points
def getEulerDis_sqr(a, b):
   return (a[0]-b[0])*(a[0]-b[0])+(a[1]-b[1])*(a[1]-b[1])

def getEulerDis_sqrt(a, b):
  return math.sqrt((a[0]-b[0])*(a[0]-b[0])+(a[1]-b[1])*(a[1]-b[1]))

#get the most likely centroid among the given point set
def getCentroid(Points):
    n = len(Points)
    tot_x = 0
    tot_y = 0
    for point in Points:
        tot_x += point[0]
        tot_y += point[1]
    tot_x /= n
    tot_y /= n
    c = (tot_x, tot_y)
    tmp_dis = sys.float_info.max
    tmp_center = ()
    for  point in Points:
        if getEulerDis_sqr(point, c) <= tmp_dis:
                tmp_dis = getEulerDis_sqr(point, c)
                tmp_center  = point
    return tmp_center

# check if two given point sets are the same
def checkChange(P, PP):
    for p in P:
        if p not in PP:
            return 0
    for p in PP:
        if p not in P:
            return 0
    return 1

# find nearest point around given x,y
def findNearestPoint(gx, gy, pointset):
  gp = (gx, gy)
  dis_min_p = 0
  dis_min = sys.float_info.max
  for p in pointset:
    if getEulerDis_sqr(p, gp) <= dis_min:
      dis_min = getEulerDis_sqr(p, gp)
      dis_min_p = p
  return dis_min_p

#the monolith to address the problem
def KMeans(numberOfClusters, x_All, timesOfIteration):
    allpoints = []
    startpoints = []
    index_ = 0
    for i in x_All:
      ti = tuple(i)
      allpoints.append(ti)

    remainpoints = allpoints.copy()

    #find perfect initial startpoints
    index_remainpoints = []
    index_startpoints = []
    for ind in range(0,len(remainpoints)):
      index_remainpoints.append(ind)

    for index_ in range(0, numberOfClusters):

          if(index_ == 0):
                index_startpoints.append(0)
                index_remainpoints.remove(0)
                continue

          outterp_details = (0,0)
          for index_point in index_remainpoints:
                  innerp_details = (0,0)
                  shortest_p_dis = sys.float_info.max
                  for index_ipoint in index_startpoints:
                        tmp_p_dis = getEulerDis_sqr(remainpoints[index_point], remainpoints[index_ipoint])
                        if shortest_p_dis > tmp_p_dis:
                            shortest_p_dis = tmp_p_dis
                  innerp_details = (shortest_p_dis, index_point)
                  if (innerp_details[0] > outterp_details[0]):
                        outterp_details = (innerp_details[0], innerp_details[1])
          if(index_ != 0):
                index_startpoints.append(outterp_details[1])
                index_remainpoints.remove(outterp_details[1])

    ilil = []
    # print(index_startpoints)

    for ind in index_startpoints:
      ilil.append(remainpoints[ind])

    # drawPointArray(ilil)

    NewCenterPoints = []
    for iter_time in range(0, timesOfIteration):
        cluster_0 = []
        cluster_1 = []
        cluster_2 = []
        cluster_3 = []
        cluster_4 = []
        cluster_5 = []
        cluster_6 = []
        cluster_7 = []
        cluster_8 = []
        cluster_9 = []
        c_0 = ()
        c_1 = ()
        c_2 = ()
        c_3 = ()
        c_4 = ()
        c_5 = ()
        c_6 = ()
        c_7 = ()
        c_8 = ()
        c_9 = ()

        tmp_ci = 0
        tmp_dis = sys.float_info.max
        for point in allpoints:
            for index in range(0, numberOfClusters):
                if getEulerDis_sqr(point, remainpoints[index_startpoints[index]]) < tmp_dis:
                    tmp_dis = getEulerDis_sqr(point, remainpoints[index_startpoints[index]])
                    tmp_ci  = index;
            if tmp_ci == 0:
                cluster_0.append(point)
            if tmp_ci == 1:
                cluster_1.append(point)
            if tmp_ci == 2:
                cluster_2.append(point)
            if tmp_ci == 3:
                cluster_3.append(point)
            if tmp_ci == 4:
                cluster_4.append(point)
            if tmp_ci == 5:
                cluster_5.append(point)
            if tmp_ci == 6:
                cluster_6.append(point)
            if tmp_ci == 7:
                cluster_7.append(point)
            if tmp_ci == 8:
                cluster_8.append(point)
            if tmp_ci == 9:
                cluster_9.append(point)
            tmp_ci = 0
            tmp_dis = sys.float_info.max
        NewCenterPoints = []
        if len(cluster_0) != 0:
            c_0 = getCentroid(cluster_0)
            NewCenterPoints.append(c_0)
        if len(cluster_1) != 0:
            c_1 = getCentroid(cluster_1)
            NewCenterPoints.append(c_1)
        if len(cluster_2) != 0:
            c_2 = getCentroid(cluster_2)
            NewCenterPoints.append(c_2)
        if len(cluster_3) != 0:
            c_3 = getCentroid(cluster_3)
            NewCenterPoints.append(c_3)
        if len(cluster_4) != 0:
            c_4 = getCentroid(cluster_4)
            NewCenterPoints.append(c_4)
        if len(cluster_5) != 0:
            c_5 = getCentroid(cluster_5)
            NewCenterPoints.append(c_5)
        if len(cluster_6) != 0:
            c_6 = getCentroid(cluster_6)
            NewCenterPoints.append(c_6)
        if len(cluster_7) != 0:
            c_7 = getCentroid(cluster_7)
            NewCenterPoints.append(c_7)
        if len(cluster_8) != 0:
            c_8 = getCentroid(cluster_8)
            NewCenterPoints.append(c_8)
        if len(cluster_9) != 0:
            c_9 = getCentroid(cluster_9)
            NewCenterPoints.append(c_9)


        if checkChange(startpoints, NewCenterPoints):
            print ('iteration time:', iter_time)
            ol=[]
            if len(cluster_0) != 0:
                il0=cluster_0
                ol.append(il0)
            if len(cluster_1) != 0:
                il1=cluster_1
                ol.append(il1)
            if len(cluster_2) != 0:
                il2=cluster_2
                ol.append(il2)
            if len(cluster_3) != 0:
                il3=cluster_3
                ol.append(il3)
            if len(cluster_4) != 0:
                il4=cluster_4
                ol.append(il4)
            if len(cluster_5) != 0:
                il5=cluster_5
                ol.append(il5)
            if len(cluster_6) != 0:
                il6=cluster_6
                ol.append(il6)
            if len(cluster_7) != 0:
                il7=cluster_7
                ol.append(il7)
            if len(cluster_8) != 0:
                il8=cluster_8
                ol.append(il8)
            if len(cluster_9) != 0:
                il9=cluster_9
                ol.append(il9)

            return ol
        if iter_time == timesOfIteration - 1:
            print ('iteration time:', iter_time)
            ol=[]
            if len(cluster_0) != 0:
                il0=cluster_0
                ol.append(il0)
            if len(cluster_1) != 0:
                il1=cluster_1
                ol.append(il1)
            if len(cluster_2) != 0:
                il2=cluster_2
                ol.append(il2)
            if len(cluster_3) != 0:
                il3=cluster_3
                ol.append(il3)
            if len(cluster_4) != 0:
                il4=cluster_4
                ol.append(il4)
            if len(cluster_5) != 0:
                il5=cluster_5
                ol.append(il5)
            if len(cluster_6) != 0:
                il6=cluster_6
                ol.append(il6)
            if len(cluster_7) != 0:
                il7=cluster_7
                ol.append(il7)
            if len(cluster_8) != 0:
                il8=cluster_8
                ol.append(il8)
            if len(cluster_9) != 0:
                il9=cluster_9
                ol.append(il9)
            return ol

        startpoints = []
        startpoints = NewCenterPoints


def drawPointCluster(pointClusterList):
  colorList=['cyan','blue','red','grey','black','yellow','green','magenta']

  for i in range(len(pointClusterList)):
    p0 = np.array([pointClusterList[i][0][0],pointClusterList[i][0][1]])
    for j in range(1,len(pointClusterList[i])):
        tmp = np.array([pointClusterList[i][j][0],pointClusterList[i][j][1]])
        p0 = np.vstack((p0,tmp))
    plt.scatter(p0[:,0],
           p0[:,1],
           marker='o',
           edgecolor='black')
#            color=colorList[i])


  plt.show()

def drawPointArray(pointArray):
    colorList=['cyan','blue','red','grey','black','yellow','green','magenta']

    ph1 =  []
    ph2 = []
    for sample in pointArray:
            ph1.append(sample[0])
            ph2.append(sample[1])
    p0 = np.array([ph1[0], ph2[0]])
    for i in range(1, len(pointArray)):
        tmp = np.array([ph1[i], ph2[i]])
        p0 = np.vstack((p0,tmp))
    plt.scatter(p0[:,0],
                        p0[:,1],
                       marker='o',
                       edgecolor='black')
#            color=colorList[i])
#   plt.xlim(-10,10)
#   plt.ylim(-11,4)

    plt.show()












###################################
x,y = make_blobs(n_samples=1350,
                n_features=2,
                centers=5,
                cluster_std=0.5,
                shuffle=True,
                random_state=9)


plt.scatter(x[:,0],
           x[:,1],
           marker='o',
           edgecolor='black')
plt.show()





ooll=KMeans(5, x, 50)

print(len(ooll))

drawPointCluster(ooll)





```
